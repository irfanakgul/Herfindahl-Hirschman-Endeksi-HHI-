```python
def fn_perc(df,var):
    #     perc 10 and 90
    # print(df.columns)
    
    perc_10_var = df[var].quantile(.1)
    perc_90_var = df[var].quantile(.9)

    div = perc_90_var - perc_10_var
    div = div / 9
    
    b_01 = perc_10_var
    b_02 = perc_10_var + div
    b_03 = b_02 + div
    b_04 = b_03 + div
    b_05 = b_04 + div
    b_06 = b_05 + div
    b_07 = b_06 + div
    b_08 = b_07 + div
    b_09 = b_08 + div
    b_10 = b_09 + div
    
    print(f'perc_10_var={b_01},\n b02={b_02},\nb03={b_03},\nb04={b_04},\nb05={b_05},\nb06={b_06},\nb07={b_07},\nb08={b_08},\nb09={b_09},\nperc_90_var={b_10}')
    
    return b_01,b_02,b_03,b_04,b_05,b_06,b_07,b_08,b_09,b_10
```

```python
def floor_bucket(df,var1,b_01,b_02,b_03,b_04,b_05,b_06,b_07,b_08,b_09,b_10):
    var_bucket = f'{var1}_bucket'
    df[f'{var1}_bucket'] = np.where(
   
        df[var1]<=b_01,'B01',
        np.where((df[var1]>b_01) & (df[var1]<=b_02),'B02',
                 np.where((df[var1]>b_02) & (df[var1]<=b_03),'B03',
                          np.where((df[var1]>b_03) & (df[var1]<=b_04),'B04',
                                   np.where((df[var1]>b_04) & (df[var1]<=b_05),'B05',
                                           np.where((df[var1]>b_05) & (df[var1]<=b_06),'B06',
                                                     np.where((df[var1]>b_06) & (df[var1]<=b_07),'B07',
                                                             np.where((df[var1]>b_07) & (df[var1]<=b_08),'B08',
                                                                      np.where((df[var1]>b_08) & (df[var1]<=b_09),'B09',
                                                                               np.where((df[var1]>b_09) & (df[var1]<=b_10),'B10',
                                                                                         'B11' ))))))))))
    
    
    buckets = ["B01", "B02", "B03", "B04", "B05", "B06", "B07", "B08", "B09", "B10", "B11"]
    no_exist_bucket =[]
    for b in buckets:
        if not b in df[var_bucket].unique().tolist():
            no_exist_bucket.append(b)
                                                                       
   
    return df,no_exist_bucket
```

```python
def fn_HHI_test(df_all,var1, var2=None):
#     print(df.columns)
    
    df = df_all[(df_all['measurementPeriodId']>=202001) & (df_all['measurementPeriodId']<=202012)]
    df_dev = df_all[(df_all['measurementPeriodId']>=201001) & (df_all['measurementPeriodId']<=201612)]
    
    # december cohorts
    df = df[df['month']==12]
    
    b_01,b_02,b_03,b_04,b_05,b_06,b_07,b_08,b_09,b_10 = fn_perc(df_dev,var1)
    df_buck_var1,no_exist_bucket = floor_bucket(df,var1,b_01,b_02,b_03,b_04,b_05,b_06,b_07,b_08,b_09,b_10)
    df_buck_var1.drop([var1], axis=1, inplace=True)
       
    var1_bucket = f'{var1}_bucket'
    
    df_var1 = pd.crosstab(df_buck_var1[var1_bucket].fillna('missing'), df_buck_var1['month'].fillna('missing'),
                          margins=False, dropna=False, normalize='columns')
    
    all_buckets = ["B01", "B02", "B03", "B04", "B05", "B06", "B07", "B08", "B09", "B10", "B11"]
    
    if len(no_exist_bucket)!=0:
        all_buckets = [i for i in all_buckets if not i in no_exist_bucket]
    
    outcome = HHI(
    actual_model=df_var1.iloc[:,0],
    bucket_names=all_buckets,
    thresholds={Threshold.GREEN: 0.001, Threshold.RED: 0.025},
    normalize = True
    ).compare()
    
    if outcome.normalized_hhi < 0.1:
        status1 = 'Green'
    elif 0.1 < outcome.normalized_hhi < 0.25:
        status1 = 'Amber'
    elif outcome.normalized_hhi >= 0.25:
        status1= 'Red'
    
    dic = {
        
        'Product':df.eadSegment.unique(),
        'EadProduct':var1,
        'normalized_hhi_ead':float("{:.2f}".format(outcome.normalized_hhi)),
        'Traffic_light1':status1}
    
    result_hhi = pd.DataFrame(dic,index=[0])
    
    ############################################ var 2 ################## 
    
    if var2 is not None:
        b_01,b_02,b_03,b_04,b_05,b_06,b_07,b_08,b_09,b_10 = fn_perc(df_dev,var2)
        df_buck_var2,no_exist_bucket2 = floor_bucket(df,var2,b_01,b_02,b_03,b_04,b_05,b_06,b_07,b_08,b_09,b_10)
        df_buck_var2.drop([var2], axis=1,inplace=True)

        var2_bucket = f'{var2}_bucket'

        df_var2 = pd.crosstab(df_buck_var2[var2_bucket].fillna('missing'), df_buck_var2['month'].fillna('missing'),
                              margins=False, dropna=False, normalize='columns')
        
        if len(no_exist_bucket2)!=0:
        
            all_buckets = [i for i in all_buckets if not i in no_exist_bucket2]

        outcome2 = HHI(
        actual_model=df_var2.iloc[:,0],
        bucket_names=all_buckets,
        thresholds={Threshold.GREEN: 0.001, Threshold.RED: 0.025},
        normalize = True
        ).compare()

        if outcome.normalized_hhi < 0.1:
            status2 = 'Green'
        elif 0.1 < outcome.normalized_hhi < 0.25:
            status2 = 'Amber'
        elif outcome.normalized_hhi >= 0.25:
            status2= 'Red'

        dic2 = {
            'Product':df.eadSegment.unique(),
            'Model_Parameter':var2,
            'normalized_hhi_mp':float("{:.2f}".format(outcome2.normalized_hhi)),
            'Traffic_light2':status2}

        result_hhi2 = pd.DataFrame(dic2,index=[0])

        result_hhi = pd.merge(result_hhi,result_hhi2, on='Product')

    return result_hhi
    
```
